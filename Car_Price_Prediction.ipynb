```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/carprices.csv')
df
```

```python
import seaborn as sns
for col in df.columns:
    sns.scatterplot(x = df[col], y = df['Sell Price($)'])
    plt.title(f'{col} vs sell price' )
    plt.xlabel(col)
    plt.ylabel("sell Price")
    plt.show()
```

Creating dummy variables using pandas

```python
dummies = pd.get_dummies(df['Car Model']).astype(int)
dummies
```

```python
merged_dummy = pd.concat([df,dummies],axis = 'columns')
merged_dummy
```

```python
new_df = merged_dummy.drop(['Car Model'], axis = 'columns')
new_df
```

```python
new_df = new_df.drop(['Mercedez Benz C class'], axis = 'columns')
new_df
```

```python
X_train = new_df.drop(['Sell Price($)'],axis = 'columns')
X_train
```

```python
Y = new_df['Sell Price($)']
Y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,Y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(X_train,Y)
```
